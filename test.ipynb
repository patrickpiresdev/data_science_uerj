```python
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
df.sort_values(by = 'name')
```

```python
duplicated = df.duplicated(subset=['name', 'age'], keep=False)
df[duplicated].sort_values(by = 'name')
```

```python
df.drop_duplicates(subset=['name', 'age'], keep='first').sort_values(by = 'name')
```

```python
df.groupby(by = ['name', 'age', 'city']).agg({'height': 'mean', 'weight': 'mean'}).reset_index()
```
